```python
#BITPLAN FROM INTERNET
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

def bit_plane_slicing(image, bit_plane):
    # Estrai il piano di bit desiderato
    return np.bitwise_and(image, 1 << bit_plane) >> bit_plane

image_np = np.array(gray)

# Visualizza i piani di bit da 0 a 7
plt.figure(figsize=(12, 6))
for i in range(8):
    plt.subplot(2, 4, i + 1)
    plt.title(f"Bit Plane {i}")
    bit_plane_image = bit_plane_slicing(image_np, i)
    plt.imshow(bit_plane_image, cmap='gray')
    plt.axis('off')

plt.tight_layout()
plt.show()

```

```python
# BITPLAN 
def bitplanna(n):
    array= gray.copy()
    h=array.shape[0]
    w=array.shape[1]
    array2 = np.zeros((h, w), dtype= 'S8')
    for a in range(h):
        for b in range(w):
            q=array[a,b]
            array2[a,b]= format(q,'08b')[n]

    array3 = np.vectorize(lambda x: (x.decode('utf-8')=='1'))(array2)
    plt.imshow(array3, cmap="gray")

bitplanna(1)
```

```python
#SCOMPONGO LA MATRICE IN RGB E CREO LE TRE MATRICI R G E B

import numpy as np
#array = np.random.randint(0, 256, size=(100, 100, 3))
#arrayP= array.copy()
array = Deconstruct.copy()
arrayP= array.copy()

print(arrayP.shape)
w=arrayP.shape[1]
h=arrayP.shape[2]

RGB=np.zeros((w,h), dtype=np.int8)
print(RGB.shape)

Red= np.zeros((w,h), dtype=np.int8)
Green= np.zeros((w,h), dtype=np.int8)
Blue = np.zeros((w,h), dtype=np.int8)

for a in range(w):
    if a%2==0:
        for b in range(h):
            if b%2==0:
                RGB[a,b]=(arrayP[0,a,b])
                Red[a,b]=(arrayP[0,a,b])
                Green[a,b]=(-1)
                Blue[a,b]=(-1)
            else:
                RGB[a,b]=(array[1,a,b])
                Green[a,b]=(arrayP[1,a,b])
                Blue[a,b]=(-1)
                Red[a,b]=(-1)
                
    else:
        for b in range(h):
            if b%2!=0:
                RGB[a,b]=(arrayP[2,a,b])
                Blue[a,b]=(arrayP[2,a,b])
                Red[a,b]=(-1)
                Green[a,b]=(-1) 
            else:
                RGB[a,b]=(array[1,a,b])
                Green[a,b]=(arrayP[1,a,b])
                Blue[a,b]=(-1)
                Red[a,b]=(-1)
```

```python
#FOURIER
import numpy as np
import cv2
from matplotlib import pyplot as plt
# Load an image in grayscale
img = array[:,:,0]
# Perform DFT and shift the result to center
dft = cv2.dft(np.float32(img), flags=cv2.DFT_COMPLEX_OUTPUT)
dft_shift = np.fft.fftshift(dft)
# Convert the complex values to magnitude and display the image
magnitude_spectrum = 20 * np.log(cv2.magnitude(dft_shift[:,:,0], dft_shift[:,:,1]))
plt.imshow(magnitude_spectrum, cmap='gray')
plt.title('Magnitude Spectrum')
plt.show()
```

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
# Load an image in grayscale
img = array[:,:,0]
# Apply 2D FFT and shift the frequencies
f_transform = np.fft.fft2(img)
f_shift = np.fft.fftshift(f_transform)
# Obtain the magnitude spectrum and plot it
magnitude_spectrum = 20 * np.log(np.abs(f_shift))
plt.imshow(magnitude_spectrum, cmap='gray')
plt.title('Magnitude Spectrum')
plt.show()
```
