1. merge(im1, im2, tp, ptOut): taken as input two images (im1, im2) the method returns a new image consisting 
of half im1 and half im2. The merge occurs according to one of the following types defined by tp (this value is 
chosen by the operator before calling the merge method): 

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
#PREPROCESSING
im1= cv2.imread(r"C:\Users\andre\Pictures\237922.jpg")
im1= cv2.cvtColor(im1, cv2.COLOR_BGR2RGB)
im1 = cv2.resize(im1, (759, 946), interpolation=cv2.INTER_LINEAR)

im2= cv2.imread(r"C:\Users\andre\Pictures\237920.jpg")
im2= cv2.cvtColor(im2, cv2.COLOR_BGR2RGB)
plt.imshow(im1)
print(im1.shape, im2.shape)
im1.shape
if im1.shape[0]%2==0:
    h=im1.shape[0]
else:
    h=im1.shape[0]-1
if im1.shape[1]%2==0:
    w=im1.shape[1]
else:
    w=im1.shape[1]-1
im1=im1[:h,:w,:]
im2=im2[:h,:w,:]
print(im1.shape, im2.shape)
```

```python
def merge(im1, im2, tp, ptOut):
    if im1.shape[0]%2==0:
        h=im1.shape[0]
    else:
        h=im1.shape[0]-1
    if im1.shape[1]%2==0:
        w=im1.shape[1]
    else:
        w=im1.shape[1]-1
    im1=im1[:h,:w,:]
    im2=im2[:h,:w,:]
    if tp==0:
        tp=im1.copy()
        #tp0[:,:w//2,:]=im1[:,:w//2,:]
        tp[:,w//2:w,:]=im2[:,w//2:w,:]
    elif tp==1:
        tp=im1.copy()
        #tp1[:h//2,:,:]=im1[:h//2,:,:]
        tp[h//2:h,:,:]=im2[h//2:h,:,:]
    elif tp==2:
        tp=im1.copy()
        tp[:h//2,:,:]=im2[:h//2,:,:]
        #tp1[h//2:h,:,:]=im1[h//2:h,:,:]
    elif tp==3:
        tp=im1.copy()
        tp[:,:w//2,:]=im2[:,:w//2,:]
        #tp3[:,w//2:w,:]=im1[:,w//2:w,:]
    plt.imshow(tp)
    cv2.imwrite(ptOut, tp)

merge(im1, im2,0, 'HalfandHalf.jpg')
```

2. conv(im, coordinates, kernel, ptOut): the method returns a ne w image to which a convolution operation is 
applied only in the region defined in coordinates. In addition, the method, will save the image in a predefined 
directory (ptOut) with a new name chosen specifically by the user (the operation of putting the name must be 
provided within the method).  

```python
imconv=im1.copy()
kernel=[[-2, -1, 0], [-1,  1, 1], [ 0,  1, 2]]

def conv(im, w0,w1,h0,h1,kernel, ptOut):
    w0=100
    w1=400
    h0=200
    h1=500

    kernel=[[-2, -1, 0], [-1,  1, 1], [ 0,  1, 2]]
    #a = np.ones((3, 3), np.float32) / 9  
    a=kernel

    conv1=imconv[h0-1:h1+1,w0-1:w1+1,:].copy()

    plt.imshow(conv1)

    imC= np.zeros((conv1.shape[0], conv1.shape[1], 3), dtype='uint8')

    for canal in range(3):
        for i in range(1,conv1.shape[0]-1):
            for j in range(1,conv1.shape[1]-1):
                imC[i,j,canal]= np.sum(conv1[i-1:i+2,j-1:j+2,canal]*a)

    print(imC.shape)
    imC=imC[1:-1,1:-1,:]
    print(imC.shape)
    imconv[h0:h1,w0:w1,:]=imC
    plt.imshow(imconv)
    cv2.imwrite(ptOut, imconv)



conv(im1, w0,w1,h0,h1,kernel,"conv.jpg")
```
