Scrivi un programma che prenda un'immagine come input, la suddivida in sezioni di dimensione nxn pixel e riorganizzi queste sezioni in ordine casuale, mischiandole in modo randomico per creare un'immagine finale mescolata.

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import random
```

```python
img= cv2.imread(r"C:\Users\andre\Pictures\15_0.png")
img= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img=img[89:,:,:]
plt.imshow(img)

```

```python
n=5
img=img[0:img.shape[0]//n*n,0:img.shape[1]//n*n,:]
plt.imshow(img)
```

```python
img1=img.copy()
w=img.shape[0]//n
h=img.shape[1]//n
print(w,h)
```

```python
A = [(x, y) for x in range(0,img.shape[0],n) for y in range(0,img.shape[1],n)]
random.shuffle(A)
#print(A)
B = A.copy()
B=B[::-1]
#random.shuffle(B)

D=[]
for a in range(0,img.shape[0],n):
    for b in range(0,img.shape[1],n):
        C=A.pop()
        x=C[0]
        y=C[1]
        img1[a:a+n,b:b+n]=img[x:x+n,y:y+n]
        #D.append(((x-a)*-1,(y-b)*-1))


D = [(x, y) for x in range(0,img.shape[0],n) for y in range(0,img.shape[1],n)]

```

```python
print(B[3][1])
print(B[3][0])

```

```python
#COMING BAAAAACK!!!!!!
img2=img1.copy()
for a in range(0,img.shape[0],n):
    for b in range(0,img.shape[1],n):
        C=B.index((a,b))
        x=D[C][0]
        y=D[C][1]
        img2[a:a+n,b:b+n]=img1[x:x+n,y:y+n]
```

```python
#Scacchiera positivo e negativo
d=0
for a in range(0,img.shape[0],n):
    for b in range(0,img.shape[1],n):
        if d%2==0:
            img1[a:a+n,b:b+n]=img[a:a+n,b:b+n]
        else:
            img1[a:a+n,b:b+n]=255-img[a:a+n,b:b+n]
        d+=1
    d+=1

plt.imshow(img1[:,:400,:])

```
