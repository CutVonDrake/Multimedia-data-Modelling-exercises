```python
#APRIRE E GESTIRE VIDEO

import cv2

d=0
a=1

# Specifica il percorso del file video
video_path = r"C:\Users\andre\Videos\Captures\War - Personale - Microsoft Edge 2023-01-13 17-09-55.mp4"  # Sostituisci con il percorso del tuo video
cap = cv2.VideoCapture(video_path)

print("Premi 'q' per uscire dalla riproduzione.")
# Specifica un fattore di scala per il ridimensionamento (es. 0.5 per il 50% delle dimensioni originali)
resize_scale = 0.5

while True:
    # Legge un fotogramma dal file video
    ret, frame = cap.read()
    #Qui sta leggendo il fotogramma quindi per modificare l'immagine fare tutto qui
    
    # Ridimensiona il fotogramma
    frame = cv2.resize(frame, None, fx=resize_scale, fy=resize_scale, interpolation=cv2.INTER_AREA)
    img=frame
    
    random.seed(42)
    n=5
    img=img[0:img.shape[0]//n*n,0:img.shape[1]//n*n,:]
    img1=img.copy()
    A = [(x, y) for x in range(0,img.shape[0],n) for y in range(0,img.shape[1],n)]
    random.shuffle(A)
    D=[]
    for a in range(0,img.shape[0],n):
        for b in range(0,img.shape[1],n):
            C=A.pop()
            x=C[0]
            y=C[1]
            img1[a:a+n,b:b+n]=img[x:x+n,y:y+n]

    frame=img1



    # Se non ci sono pi fotogrammi, esci dal ciclo
    if not ret:
        print("Fine del video.")
        break

    # Mostra il fotogramma
    #cv2.imshow('Riproduzione video', frame)

    # Percorso della cartella in cui vuoi salvare l'immagine
    output_folder = r'C:\Users\andre\Python notebooks\Media\frames'
    # Nome del file di output
    d+=1
    output_file = os.path.join(output_folder, f'saved_image{d}.jpg')
    # Salva l'immagine
    cv2.imwrite(output_file, frame)

    # Esci dal ciclo premendo 'q'
    if cv2.waitKey(30) & 0xFF == ord('q'):  # 30 ms di ritardo tra i frame
        break
#print(frame)
# Rilascia il video e chiudi tutte le finestre
cap.release()
cv2.destroyAllWindows()
print(a)
```

```python
import cv2
import os

# Cartella contenente i frame
frames_folder = r"C:\Users\andre\Python notebooks\Media\frames"

# Percorso di output per il video
output_video_path = r"C:\Users\andre\Python notebooks\Media\video\video.mp4"

# Leggi tutti i file nella cartella e ordina per nome
frames = sorted([f for f in os.listdir(frames_folder) if f.endswith(('.png', '.jpg', '.jpeg'))])

# Usa il primo frame per ottenere dimensioni del video
first_frame_path = os.path.join(frames_folder, frames[0])
frame = cv2.imread(first_frame_path)
height, width, layers = frame.shape

# Specifica codec e crea il VideoWriter
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec per file MP4
fps = 30  # Fotogrammi per secondo
video = cv2.VideoWriter(output_video_path, fourcc, fps, (width, height))

# Itera su tutti i frame e aggiungili al video
for frame_name in frames:
    frame_path = os.path.join(frames_folder, frame_name)
    frame = cv2.imread(frame_path)
    
    # Verifica che il frame sia valido
    if frame is None:
        print(f"Errore nel leggere {frame_name}, saltato!")
        continue
    
    # Aggiungi il frame al video
    video.write(frame)

# Rilascia il writer e chiudi
video.release()
print(f"Video salvato in: {output_video_path}")
```

```python
#Visualizzare i frame salvati come video


# Percorso della cartella contenente i frame
frames_folder = r'C:\Users\andre\Python notebooks\Media\frames'
fps = 30

# Ottieni la lista dei file nella cartella e ordina i nomi
frame_files = (os.listdir(frames_folder))  # Ordine alfabetico


# Riproduzione dei frame
for frame_file in frame_files:
    # Costruisci il percorso completo del frame
    frame_path = os.path.join(frames_folder, frame_file)
    
    # Leggi il frame
    frame = cv2.imread(frame_path)
    
    if frame is None:
        print(f"Impossibile leggere il frame: {frame_file}")
        continue
    
    # Mostra il frame
    cv2.imshow('Video', frame)
    
    # Attendi un periodo in base agli FPS
    if cv2.waitKey(int(1000 / fps)) & 0xFF == ord('q'):
        break  # Esci premendo 'q'

# Chiudi la finestra di visualizzazione
cv2.destroyAllWindows()
```
