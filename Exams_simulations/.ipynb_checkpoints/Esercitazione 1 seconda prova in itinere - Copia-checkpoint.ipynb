```python
bitplane(im, t, ptOut): the method takes as input in image and an integer value t. If t == 0 then all bit planes of 
odd position will be set to 0. If t == 1 then all even position bit planes will be set to 0. In addition, the method, 
will save the image in a predefined directory (ptOut) with a new name chosen specifically by the user (the 
operation of putting the name must be provided within the method). 
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(r"C:\Users\andre\Pictures\237922.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)


binar = np.vectorize(lambda x: format(x, 'b').zfill(8))(img)
#print(binar)
```

```python
t=0
for z in range(3):
    for a in range(binar.shape[0]):
        for b in range(binar.shape[1]):
            bin2=''
            for c in range(8):
                if t==0:
                    if c%2!=0:
                        bin2+='0'
                    else:
                        bin2+=binar[a,b,z][c]
                        
                elif t==1:
                    if c%2==0:
                        bin2+='0'
                    else:
                        bin2+=binar[a,b,z][c]
            binar[a,b,z]=bin2


img_reversed = np.vectorize(lambda x: int(x, 2))(binar)




                        
                        
```
