```python
import pandas as pd
import numpy as np
import math
```

```python
splits = {'train': 'plain_text/train-00000-of-00001.parquet', 'test': 'plain_text/test-00000-of-00001.parquet', 'unsupervised': 'plain_text/unsupervised-00000-of-00001.parquet'}
dfC = pd.read_parquet("hf://datasets/stanfordnlp/imdb/" + splits["train"])
```

```python
import nltk
#nltk.download('punkt_tab')
#nltk.download('punkt')
from nltk.tokenize import word_tokenize as wt
```

```python
V=set()
docs= [ [],[] ]
for i, sample in df.iterrows():
    tokens=wt(sample['text'])
    V= V.union(set(tokens))
    docs[sample['label']].extend(tokens) # extend per evitare liste annidate
```

```python
count_w_c = [{w:0 for w in V} for c in range(2)]

for c in range(2):
    for x in docs[c]:
        count_w_c[c][x]+=1
```

```python
LogPwc= [{w: math.log10(count_w_c[c][w] + 1 /(len(docs[c]) + len(V))) for w in V} for c in range(2)]

print(LogPwc)
```
